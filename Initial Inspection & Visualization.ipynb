## The shift() method

```python
close.head()
```

```python
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.head()
```

```python
aapl.shift(periods = 1) # The period will shift the value to the next row or the previous row if you choose a -ve number
```

```python
aapl["lag1"] = aapl.shift(periods = 1)
```

```python
aapl.head()
```

```python
aapl.AAPL.sub(aapl.lag1)
```

```python
aapl["Diff"] = aapl.AAPL.sub(aapl.lag1)
```

```python
aapl.head()
```

```python
aapl.AAPL.div(aapl.lag1).sub(1).mul(100) # Calculate Daily Percentage Change 
```

```python
aapl["pct_change"] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
aapl.head()
```

## Financial Time Series - Returns and Risks 

```python
import numpy as np
```

```python
norm.plot(figsize = (15,8), fontsize = 13)
plt.legend(fontsize=13)
plt.show()
```

```python
close.head()
```

```python
close.pct_change().dropna()
```

```python
ret = close.pct_change().dropna()
```

```python
ret.head()
```

```python
summary = ret.describe().T.loc[:,["mean","std"]]
summary
```

```python
summary["mean"] = summary["mean"] * 252
summary["std"] = summary["std"] * np.sqrt(252)
```

```python
summary
```

```python
summary.plot.scatter(x = "std", y ="mean", figsize=(12,8), s= 50, fontsize = 15)
#labels with ticker symbol with FOR LOOP
for i in summary.index: 
     plt.annotate(i, xy =(summary.loc[i,"std"]+0.002, summary.loc[i, "mean"]+0.002), size = 15)
plt.xlabel("ann. Risk(std)", fontsize=15)
plt.ylabel("ann. Return", fontsize = 15)
plt.title("Risk/Return", fontsize = 20)
plt.show()
```

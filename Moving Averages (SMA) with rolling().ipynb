## Momentum Trading Strategies 

```python
SP500.head()
```

```python
SP500.tail()
```

```python
SP500["SMA50"] = SP500.rolling(window = 50, min_periods = 50).mean()
```

```python
SP500
```

```python
SP500.plot(figsize=(12,8), fontsize = 15)
plt.legend(loc = "upper left", fontsize = 15)
plt.show()
```

```python
SP500["SMA200"] = SP500.Close.rolling(window = 200, min_periods = 200).mean()
```

```python
SP500.tail()
```

```python
SP500.info()
```

```python
SP500.plot(figsize=(12,8), fontsize=15)
plt.legend(fontsize = 15)
plt.show()
```

```python
# Only Plot The last two columns
SP500.iloc[:,-2:].plot(figsize=(12,8), fontsize=15)
plt.legend(fontsize = 15)
plt.show()
# Note:
# The trader invests when the shorter term SMA is above the longer term SMA and even goes short when the opposite holds true
# The rationale behind this is that shorter SMA catches the most recent events 
```

## Exponetionally weighted moving averages 

```python
SP500 = pd.read_csv("SP500.csv", parse_dates= ["Date"], index_col = "Date", usecols= ["Date", "Close"])
SP500
```

```python
SP500 = SP500.loc["2008-12-31":"2018-12-31"].copy()
```

```python
SP500.Close.rolling(window = 10).mean()
```

```python
SP500.Close.ewm(span = 10, min_periods = 10).mean()
```

```python
SP500["SMA"] = SP500.Close.rolling(window = 100).mean()
SP500["EMA"] = SP500.Close.ewm(span = 100, min_periods = 100).mean()
```

```python
SP500
```

```python
SP500.iloc[:,-2:].plot(figsize = (12,8), fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

# Importing Financial Data From Excel

```python
import pandas as pd 
```

```python
pd.read_excel("SP500.xls")
```

```python
pd.read_excel("SP500.xls", parse_dates = ["Date"], index_col = "Date")
```

```python
pd.read_excel("SP500.xls", parse_dates = ["Date"], index_col = "Date", usecols = "A:G") 
```

```python
pd.read_excel("SP500.xls", parse_dates = ["Date"], index_col = "Date", usecols = "A,B,C, E:G" )
```

```python
pd.read_excel("SP500.xls", sheet_name = "SP500") # or sheet_name = 0 as its on index position 0 
```

```python
pd.read_excel("SP500.xls", sheet_name = "Sales")
```

```python
SP500 = pd.read_excel("SP500.xls", parse_dates = ["Date"], index_col = "Date", usecols = "A:E") 
```
